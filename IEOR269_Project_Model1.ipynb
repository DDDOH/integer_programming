```python
import pandas as pd
import numpy as np
import math
from gurobipy import *
```

```python
#parameter
K=10
M=100
```

```python
anchors = pd.read_excel('a0.xls', usecols = range(26), header=0, nrows=26) 
print(anchors)
```

```python
sensors = pd.read_excel('y0.xls', usecols = range(26), header=0, nrows=26)
print(sensors)
```

```python
sensor = sensors.values
print(sensor)
J = range(len(sensor[:,1]))
print(J)
anchor= anchors.values
print(anchor)
I = range(len(anchor[:,1]))
print(I)
```

```python
dist = pd.read_excel('dist0.xls', usecols = range(26),nrows=26)
print(dist)
```

```python
distance=dist.values
print(distance)
```

```python
m = Model("Sensor Network Model 1")

#Define variables
x = m.addVars(I, vtype=GRB.BINARY, name="hide")
t = m.addVars(I, J, lb=0, name="estimation error for pair i and j")
mu=m.addVars(I, J,lb=0, name="coefficient mu in KKT")
V=m.addVars(I, J, name="V in KKT")
y_x=m.addVars(J, name="x coordinates of sensors")
y_y=m.addVars(J, name="y coordinates of sensors")
lam_x=m.addVars(I,J,name="x coordinate of lam coef")
lam_y=m.addVars(I,J,name="y coordinate of lam coef")

#这些是失败了的
#y=m.addMVar((len(sensor[:,]),2),vtype=GRB.CONTINUOUS, name="estimated location of sensors")
#lam=m.addMVar((len(anchor[:,1]),len(sensor[:,1])),vtype=GRB.CONTINUOUS)
m.update()
```

```python
#Objective function
m.modelSense = GRB.MINIMIZE
m.setObjective(quicksum((y_x[j]-sensor[j,0])*(y_x[j]-sensor[j,0])+(y_y[j]-sensor[j,1])*(y_y[j]-sensor[j,1]) for j in J))

#矩阵运算和Norm公式都不让用
#m.setObjective(quicksum(np.linalg.norm(sensor[j,:]-y[j,:]) for j in J))
#m.setObjective((y[:,0]-sensor[:,0]) @ (y[:,0]) + (y[:,1]) @ (y[:,1]))
```

```python
#y=m.addMVar((25,2),vtype=GRB.CONTINUOUS, name="estimated location of sensors")

#m.addConstr(y[:,0] @ y[:,1] +y[:,1]@ y[:,1]<=5 )
```

```python
#Constriants
m.addConstr(quicksum(x[i] for i in I)<=K)

m.addConstrs(V[i,j]+mu[i,j] * distance[i][j]==1 for i in I for j in J)

m.addConstrs(lam_x.sum('*',j)==0 for j in J)

m.addConstrs(lam_y.sum('*',j)==0 for j in J)

m.addConstrs(t[i,j] * V[i,j]<=0 for i in I for j in J)

m.addConstrs((anchor[i,0]-y_x[j])*(anchor[i,0]-y_x[j])+(anchor[i,1]-y_y[j])*(anchor[i,1]-y_y[j])
             <=(distance[i][j]*(1+t[i,j])+M * (1-x[i]))*(distance[i][j]*(1+t[i,j])+M * (1-x[i])) for i in I for j in J)

m.addConstrs(lam_x[i,j]*lam_x[i,j]+lam_y[i,j]*lam_y[i,j]<=mu[i,j]*mu[i,j] for i in I for j in J)

#下面两个是或

m.addConstrs(mu[i,j]<=M * x[i] for i in I for j in J)

m.addConstrs(lam_x[i,j]*(anchor[i,0]-y_x[j])+lam_y[i,j]*(anchor[i,1]-y_y[j])+mu[i,j]*(distance[i][j]*(1+t[i,j])+M*(1-x[i]))
             <=0 for i in I for j in J)

#norm 不让用 平方不让用 变量弄成矩阵也不让我进行矩阵运算 这TM算个锤子
#m.addConstrs(np.linalg.norm(anchor[i,:]-y[j,:])<=d[i][j]*(1+t[i][j])+M*(1-x[i]) for i in I for j in J)
#m.addConstrs(np.linalg.norm(lam[i][j])<=mu[i][j] for i in I for j in J)

m.update()
```

```python
m.optimize()
```
